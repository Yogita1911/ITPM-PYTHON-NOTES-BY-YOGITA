### Print () : The simplest way to produce output is using the print() function where you can pass zero or more expressions separated by commas. This function converts the expressions you pass into a string before writing to the screen. 


```python
#1.simple way
name = "Raj"
age = 25
print(name)
print(age)
```

```python
#2.The most common way to print values
name = "Raj"
age = 25
print("Name:", name)
print("Age:", age)
```

```python
#3.Concatenating strings with the + operator.
name = "Raj"
age = 25
print("Name: " + name + ", Age: " + str(age))
```

```python
Name = "Shubham"
Rall_no = 36
print("Student Name = " + Name+ ", Roll Number of Student:" +str(Rall_no))
```

```python
Name = "Shubham"
Address = "pune sector 60"
print("Name of Student ="+Name+ ", Address of student is " + Address)
# o/p Name of Student = shubham ,Address of student is pune
```

```python
#4.Formatted string literals for embedding expressions.(f STRING)
name = "Alice"
age = 25
print(f"Name: {name}, Age: {age}")
```

```python
Name = "Shubham"
Address = "pune sector 60"
print(f"Name of Student: {Name} , Address of Student: {Address}")
```

```python
#5.Automatically adds spaces between values.
name = "Alice"
age = 25
print("Name:", name, "Age:", age)
```

```python
Name = "Shubham"
Address = "pune sector 60"
print("Name:",Name , "Address:",Address)
```

```python
#6.Custom Separator: (,)
name = "Raj"
age = 25
print("Raj", "25" ,sep= ", ")
```

```python
Name = "Dhanajay"
Address = "pune"
Age = 45
#simple way
print(Name)
print(Address)
print(Age)
#common way
print("Name:", Name)
print("Address:",Address)
print("Age:",Age)
#string concatination
print("Student Name = " + Name+ ", Address of Student:" +Address+ ",Age :" +str(Age))
# f string
print(f"Name: {name}, Address : {Address} ,Age: {age}")
#automatically 
print("Name:", name, "Address:",Address ,"Age:", age)
#using sep operator
print("Dhananjay","pune", "25" ,sep=", ")
```
