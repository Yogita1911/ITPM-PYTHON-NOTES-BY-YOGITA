### How to edit items from list

### How to add items in list

```python
# add items using appnd()
L4= [100, 200, 300, 400, 500]
L4.append(600)
print(L4)
L4.append([700,800])
print(L4)

#add items using extend()
L4.extend([900,1000])
print(L4)


# add item using insert
L4.insert(1,"hello")
print(L4)
```

### How to delete items from list

```python
L5 = [1,2,3]
del L5
print(L5)
```

```python
# delete using del
L6 = [100, 'hello', 200, 300, 400, 500, 600, [700, 800], 900, 1000]
del L6[1]
print(L6)


#delete using remove
L6.remove(500)
print(L6)

#using pop
L6.pop()
print(L6)
 

#using clear
L6.clear()
print(L6)
```
