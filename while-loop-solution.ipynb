```python
### 5. Find the Greatest Common Divisor (GCD) of Two Numbers
num1=int(input("enter 1st number : "))
num2=int(input("enter 2nd number : "))
i=1
gcd=0
if num1==0 or num2==0:
    print("zero not allowed")
elif num1<num2:
    check=num1
else:
    check=num2

while i<=check:
    if num1%i==0:
        if num2%i==0:
            gcd=i
    i+=1
print(gcd)
```

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
gcd = 1
i = 1 
while i <= min(num1, num2):  
  if num1 % i == 0 and num2 % i == 0:  
    gcd = i  
  i += 1
print(f"The GCD of {num1} and {num2} is: {gcd}")
```

```python
# 9. Calculate the Power of a Numbers without using operator.  
#9. Calculate the Power of a Number Without Using the Operator
base = int(input("Enter the base: "))
exponent = int(input("Enter the exponent: "))
result = 1
count = 0
while count < exponent:
    result = result * base
    count = count + 1
print(f"{base} raised to the power of exponent {exponent} is {result}")
```

```python
# 11. Sum of Natural Numbers Until a Given Number
num = int(input("Enter a number: "))
sum_of_numbers = 0
i = 1
while i <= num:
    sum_of_numbers += i
    i += 1
print(f"The sum of natural numbers up to {num} is {sum_of_numbers}.")
```
