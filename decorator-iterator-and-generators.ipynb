````python
#### Syntax of a Decorator:
#1. `decorator_function` is the decorator.
#2. `display()` is the original function being decorated.
#3. The `@decorator_function` syntax is a shorthand for `display = decorator_function(display)`.

def decorator_function(original_function):
    def wrapper_function():
        print("Wrapper executed before the original function.")
        
        #print("Wrapper executed after the original function.")
        original_function()
    return wrapper_function

@decorator_function
def original_function():
    print("Original function executed.")

original_function()
````

````python
def simple_decorator(func):
    def wrapper():
        print("Well-come ")
        func()
        print("Thank you for visiting.")
    return wrapper

@simple_decorator
def func():
    print("Your balance is : 50000")

func()
````

### generator : is a special type of iterable that allows you to generate values one at a time using the yield keyword. Generators are memory-efficient because they do not store the entire sequence in memory; instead, they produce items lazily, as needed.

### Generators in Python

#### A **generator** is a special type of function that produces a sequence of results lazily, one at a time, using the `yield` keyword. Unlike regular functions that return     a, b = b, a + b





````python
for num in range(10):
    print(num)
````

````python
# Initialize the starting number
number = 1

# Use a while loop to iterate until 100
while number <= 100:
    print(number)
    number += 1  # Increment the number
````

````python
def creator(): 
    number = 1
    while number <= 100:
        yield number
        number += 1 
print(creator())
x = creator()
print(next(x))
print(next(x))
print(next(x))
````

````python
def count_up_to(limit):
    count = 1
    while count <= limit:
        yield count
        count += 1

# Usage
gen = count_up_to(10)
for number in gen:
    print(number)
````

###  iterators in Python are powerful tools for traversing sequences and collections, 
### and they can be easily customized for more advanced iteration needs.

### In Python, an **iterator** is an object that implements two methods:

1. **`__iter__()`**: Returns the iterator object itself. This is used in the case of custom iterators or objects that are iterable.
2. **`__next__()`**: Returns the next item from the collection. When there are no more items, it raises the `StopIteration` exception to signal that the iteration is complete.

Iterators allow you to loop through a collection (like a list, tuple, or dictionary) one element at a time, without needing to know the underlying structure of the collection.

 The `__init__` method initializes the starting and ending points for the iteration.
- The `__iter__` method returns the iterator object itself (`self`), making this object an iterator.
- The `__next__` method defines how the iteration happens: it returns the current value and increments it, until it reaches the end and raises `StopIteration`.


````python
# Example with iter() and next()
my_list = [10, 20, 30,40,50,60]
it = iter(my_list)

print(next(it))  # Output: 10
print(next(it))  # Output: 20
#print(next(it)) 
````

````python
# Using a built-in iterator
my_list = [1, 2, 3, 4, 5]

# Create an iterator for the list
list_iter = iter(my_list)

# Manually use the iterator to get values
print(next(list_iter))  # Output: 1
print(next(list_iter))  # Output: 2
print(next(list_iter))  # Output: 3
````

````python
import threading
import time
class A(threading.Thread):
    def run(self):
        for i in range(1,5):
            print("Sayli")
            time.sleep(1)
class B(threading.Thread):
    def run(self):
        for i in range(1,5):
            print("Rohini")
            time.sleep(1)
            
t1=A()
t2=B()


t1.start()
t2.start()
````
