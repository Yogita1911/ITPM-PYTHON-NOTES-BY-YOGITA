```python
#1.Print Numbers From 1 to 10 Using a While Loop
i=1
while i<=10:
 print(i)
 i+=1
```

```python
#2.Calculate the Sum of Digits of a Given Number
num=int(input("Enter the number:"))
sum_of_digits = 0
while num > 0:
 last_digit = num % 10
 sum_of_digits += last_digit
 num //= 10
print(sum_of_digits)
```

```python
#3.Reverse a Number Using a While Loop
num=int(input("Enter the number:"))
reverse_num=0
while num>0 :
 reverse_num = reverse_num * 10 + num % 10
 num //= 10
print(reverse_num)
```

```python
#4. Print a Countdown From 10 to 1
i=10
while i>=1:
 print(i)
 i-=1
```

```python
#5.Find the Greatest Common Divisor (GCD) of Two Numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
gcd = 1
i = 1
while i <= min(num1, num2): 
 if num1 % i == 0 and num2 % i == 0: 
   gcd = i 
 i += 1
print(f"The GCD of {num1} and {num2} is: {gcd}")
```

```python
#6.Keep Taking Input From the User Until They Enter 'exit
user = input("Enter something: ")
while user != "exit": 
 print(f"You entered: {user}")
 user = input("Enter something: ")
print("Exit the program")
```

```python
#7. Check if a Number is a Palindrom
num=int(input("Enter the number:"))
reverse_num=0
original_num= num
while num>0 :
 reverse_num = reverse_num * 10 + num % 10
 num //= 10
if reverse_num == original_num:
 print("The number is Palindrome")
else:
 print("The number is not Palindrome")
```

```python
#8. Generate the Fibonacci Sequence Until a Specified Number
limit = int(input("Enter the max limit for Fibo sequence: "))
a, b = 0, 1
while a <= limit: 
 print(a, end=" ") 
 a, b = b, a + b
print("Fibonacci Sequence:")
```

```python
#9. Calculate the power of a Number Without using the Operator.
base = int(input("Enter base number: "))
exponent = int(input("Enter exponent: "))
result = 1
count = 1 
while count <= exponent: 
 result *= base 
 count += 1
print(f"{base} raised to the power of {exponent} is: {result}")
```

```python
#11. Sum of Natural Numbers Until a Given Number.
num = int(input("Enter a number: "))
sum = 0
i = 1 
while i <= num: 
 sum += i 
 i += 1
print(f"The sum of natural numbers up to {num} is: {sum}")
```

```python
#12. Implement a Simple Menu-Driven Program Using  While Loop
while True:
    print("\nSimple Calculator Menu:")
    print("1. Add two numbers")
    print("2. Subtract two numbers")
    print("3. Multiply two numbers")
    print("4. Divide two numbers")
    print("5. Exit")
    choice = input("Enter your choice (1/2/3/4/5): ")
    if choice == '1':
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        print(f"The sum is: {num1 + num2}")
    elif choice == '2':
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        print(f"The difference is: {num1 - num2}")
    elif choice == '3':
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        print(f"The product is: {num1 * num2}")
    elif choice == '4':
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        if num2 != 0:
            print(f"The quotient is: {num1 / num2}")
        else:
            print("Error! Division by zero is not allowed.")
    elif choice == '5':
        print("Exit the program")
        break
    else:
        print("Invalid choice! Please choose a valid option (1/2/3/4/5).")
```

```python
#13. Simulate a Basic Password check that Limits to 3 Attempts.
password = "kashish1234"
attempts = 0
while attempts < 3: 
    user_input = input("Enter the password: ")
 
    if user_input == password:
        print("Password Correct")
        break
    else:
        attempts += 1
        if attempts < 3:
            print(f"Incorrect password. You have {3 - attempts}attempts left.")
        else:
            print(" Too many incorrect attempts.Access denied.")
```
