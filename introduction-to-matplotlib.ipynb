```python
## Introduction to Matplotlib and Its Uses

**Matplotlib** is a popular data visualization library in Python.
It provides tools to create a wide range of static, animated, and interactive plots.
It is highly customizable, making it ideal for creating publication-quality visualizations.

### Key Features of Matplotlib:
1. **Versatility**: Supports a variety of plots, such as line plots, bar charts, scatter plots, and histograms.
2. **Customization**: Allows users to control every aspect of the plot, including colors, fonts, and gridlines.
3. **Integration**: Works well with other Python libraries like NumPy, pandas, and SciPy.
4. **Interactive Environment**: Compatible with Jupyter Notebooks, enabling real-time plot generation and interaction.

### Common Uses of Matplotlib:
- **Exploratory Data Analysis (EDA)**: Visualizing data distributions, trends, and patterns.
- **Presentation of Results**: Creating clear and concise charts for reports and presentations.
- **Data Insights**: Helping identify outliers, correlations, and other insights.

```

### 1. **Line Plots**
Line plots are used to visualize data points connected by a continuous line, often used to show trends over time.



```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y,label='Linear Growth')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

### 2. **Bar Charts**
Bar charts are used to compare different categories or groups


```python

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.bar(categories, values, color="blue")
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.plot(x, y, color='blue')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### 3. **Scatter Plots**
Scatter plots display individual data points, often used to show relationships or correlations.


```python

x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### 4. **Histograms**
Histograms are used to represent the distribution of numerical data.

```python
import numpy as np

data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=10, color='blue', alpha=0.8)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Customizing Plots

### 1. **Adding Titles and Labels**
- Use `plt.title()` to add a title.
- Use `plt.xlabel()` and `plt.ylabel()` to label axes.

### 2. **Legends**
Legends help identify different data series in a plot.
- Add a legend using `plt.legend()` and include `label` attributes in plot functions.

### 3. **Grid**
- Use `plt.grid()` to add grid lines for better readability.


```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, label='Data Line')
plt.title('Customized Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.5)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y )
plt.show()
```

### 4. **Colors and Styles**
- Change colors with parameters like `color='green'`.
- Use line styles (`linestyle='--'`, `'-.'`, etc.) and markers (`marker='o'`, `'*'`, etc.).


```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, color='black', linestyle='--', marker='o', label='Styled Line')
plt.title('Styled Plot Example')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

```python
import pandas as pd
student = {"Id":[1,2,3,4,5],
           "Name":["Amita","Anupama","Chaitali","Darshan","Shruti"],
           "Study_hours":[2,4,5,6,1],
           "Marks":[20,40,50,60,10]}
df = pd.DataFrame(student)
df
```

### Sub_plot

```python
import matplotlib.pyplot as plt

# Create 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot on each subplot
axs[0, 0].plot([1, 2, 3], [1, 4, 9])  # Top-left
axs[0, 1].scatter([1, 2, 3], [1, 4, 9])  # Top-right
axs[1, 0].bar([1, 2, 3], [1, 4, 9])  # Bottom-left
axs[1, 1].hist([1, 2, 2, 3, 3, 3, 3], bins=3)  # Bottom-right

# Set title for each subplot
axs[0, 0].set_title("Line Plot")
axs[0, 1].set_title("Scatter Plot")
axs[1, 0].set_title("Bar Plot")
axs[1, 1].set_title("Histogram")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
