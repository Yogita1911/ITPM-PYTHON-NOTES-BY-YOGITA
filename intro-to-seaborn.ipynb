## **1. Heatmaps**
Heatmaps are used to visualize data in a matrix form, often to show correlations or intensities.


syntax for **`sns.heatmap`** in the given line:


sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5
#### 1. **`sns.heatmap`**
This function in **Seaborn** creates a heatmap visualization. It is often used to display matrices, such as correlation matrices, with color-coded values.
#### 2. **`corr`**
- **Purpose**: The input data for the heatmap.
- **In this case**: It is a correlation matrix created using `data.corr()` that shows pairwise correlations between columns in a DataFrame.
- **Type**: A 2D array-like structure (e.g., Pandas DataFrame or NumPy array).
#### 3. **`annot=True`**
- **Purpose**: Displays the actual values in each cell of the heatmap.
- **Value**: 
  - `True`: Annotates each cell with its corresponding data value.
  - `False`: No annotations are displayed.

#### 4. **`cmap='coolwarm'`**
- **Purpose**: Specifies the colormap (color gradient) for the heatmap.
- **Value**:
  - `'coolwarm'`: A diverging colormap with blue for lower values and red for higher values. It's useful for datasets with both positive and negative values.
  - Other examples: `'viridis'`, `'plasma'`, `'magma'`, `'YlGnBu'`, etc.

#### 5. **`fmt='.2f'`**
- **Purpose**: Defines the string formatting for annotations.
- **Value**:
  - `'.2f'`: Rounds the values to two decimal places (e.g., `0.25`).
  - Other formats:
    - `'.1f'`: Rounds to one decimal place.
    - `'.0f'`: Displays as integers.
#### 6. **`linewidths=0.5`**
- **Purpose**: Sets the width of the lines separating the cells in the heatmap grid.
- **Value**: 
  - `0.5`: Thin lines separate each cell.
  - Higher values create thicker grid lines.

### Additional Notes:
1. **`sns.heatmap`** automatically scales the colors based on the values in the input matrix.
2. You can add options like `linecolor` to customize the grid color.



```python


# Sample dataset
data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [2, 3, 4, 5]
})

# Compute correlation matrix
corr = data.corr()

# Create a heatmap
sns.heatmap(corr, annot=True, cmap='plasma', fmt='.1f', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

## **2. Pair Plots**
Pair plots visualize pairwise relationships between numeric variables and their distributions.

```python
import seaborn as sns
import pandas as pd

# Load dataset (e.g., Iris dataset)
from seaborn import load_dataset
iris = load_dataset('iris')

# Create pair plot
sns.pairplot(iris, hue='species', diag_kind='kde', palette='husl')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
from seaborn import load_dataset
tips = load_dataset('tips')

# Create a box plot
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')
plt.title("Box Plot of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Loan_Status', data=df)
plt.title('Loan Status Distribution')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.show()
```
