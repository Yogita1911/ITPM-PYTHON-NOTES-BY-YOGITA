## Defination:


**A generator is a special type of function that remembers its state and gives one value at a time instead of returning everything at once.It uses the yield keyword instead of return.**




## How Do Generators Work:-


**Unlike normal functions (which return a value and terminate), generators pause execution at yield and resume from the same point when called again. They are memory-efficient because they donâ€™t store the entire sequence in memory.**

```python
def my_gen():
    yield 1
    yield 2
    yield 4

x = my_gen()
print(x.__next__())
print(x.__next__())
print(x.__next__())
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

fib = fibonacci(6)
print(list(fib))  
```
