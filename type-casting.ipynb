### 2. Implicit Type Casting
#### This happens automatically when Python interprets and converts one data type to another during execution. This usually occurs when performing operations where one type can be safely promoted to another.

Data Types Interacting Through Implicit Casting:
int
float
complex
bool
Number of Combinations:
By pairing these types in operations, we get the following valid combinations:

int + float
int + complex
int + bool
float + int
float + bool
float + complex
bool + int
bool + float
bool + complex




```python
# int to float
x = 4
y = 3.4
z = x+y     # inter  z = 4.0+3.4  float + flaot
print(z)
```

```python
#int + complex
x = 4
y = 5+5j
z = x+y
print(z)
```

```python
#float + complex
x = 4.5
y = 5+5j
z = x+y
print(z)
```

```python
#bool + float
x = True
y = 4.5
z = x+y
print(z)
```

```python
#bool + int
x = True
y = 4
z = x+y
print(z)
```

```python
#bool + complex
x = False
y = 4+3j
z=x+y
print(z)
```
