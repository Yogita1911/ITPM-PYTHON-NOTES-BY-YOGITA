### Rules for naming variables in Python.

#### 1.In Python, variable names must follow certain rules to ensure proper functionality and code readability:

#### 2.Valid Characters:
1.Variable names can contain letters (both uppercase and lowercase,Combination of both, numbers) and 
2.contain underscores. 
3.Special characters, spaces, or punctuation marks are not allowed.

#### 3.Start with a Letter or Underscore: 
1.A variable name must begin with a letter (a-z, A-Z) or an underscore (_). 
2.It cannot start with a number.

#### 4.Case-Sensitive: 
1.Python is case-sensitive, so my_variable and My_Variable would be considered different variables.


#### 5.Reserved Words:
1.Avoid using Python reserved words as variable names. Examples include if, else, while, for, import, and others.

#### 6.Descriptive and Meaningful: Choose variable names that are descriptive and convey the purpose of the variable for better code readability.

#### 7.Avoid Single Underscore as the First Character: While technically allowed, using a single underscore as the first character in a variable name is often reserved for special cases (e.g., _variable). It's not recommended for regular variable names.

## **Scope of Variable :**
#### The scope of a Python variable refers to the area where we may locate it and, if necessary, access it.

### **Global Variable :**
#### Global variables are those we declare and define outside of any function body or method.
#### It can be accessed anywhere in file

### **Local Variable:**
#### Global variables are those we declare and define inside of any function body or method.
#### It can be accessed within function or accessed using function name.
